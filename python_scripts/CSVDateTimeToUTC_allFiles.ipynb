```python
# package import
import pandas as pd
```

```python
# this cell converts the event_timestamp timezones to UTC

# file name iterator
files = ['process_data_12', 'process_data_13', 'process_data_16', 'process_data_17', 'process_data_18', 
         'process_data_19 may', 'condition_data', 'condition_data_de', 'fft_data'] 

for i in files :
    
    print('\n---- ' + i + '.csv ----\n')
    
    df = pd.read_csv('../og_csv/' + i + '.csv') # taken from og_csv as this was the first level of processing
    print(df['event_timestamp'].head())
    
    # converting to datetime type
    df['event_timestamp'] = pd.to_datetime(df['event_timestamp'], format = '%d/%m/%Y %H:%M:%S') 
    
    # subtracting 5 hours & 30 minutes from the initial timestamp in IST to convert to UTC
    df['event_timestamp'] = df['event_timestamp'] - pd.Timedelta(hours = 5, minutes = 30)
    
    # converting back to initial format 
    df['event_timestamp'] = df['event_timestamp'].dt.strftime('%d/%m/%Y %H:%M:%S')
    print(df['event_timestamp'].head())
    
    # saving the file in first_modified_csv directory
    df.to_csv('first_modified_csv/'+i+'_edit.csv', index = False)
    
    print('\n---- **** ----\n')
```
